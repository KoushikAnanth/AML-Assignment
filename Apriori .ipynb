```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data=pd.read_csv('AprioriEX.csv',on_bad_lines='skip')
transactions=[]
data
```

```python
data=data.drop('Purchase Address',axis=1)
data
```

```python
data.shape
```

```python
for i in range(0,9723):
    transactions.append([str(data.values[i,j])for j in range(0,4)])
transactions    
```

```python
from apyori import apriori
rules=apriori(transactions=transactions,min_support=0.002,min_confidence=0.1,min_lift=2,min_length=2,max_length=2)
print(rules)
```

```python
results=list(rules)
results
```

```python
def inspect(results):
    lhs=[tuple(result[2][0][0])[0] for result in results]
    rhs=[tuple(result[2][0][1])[0] for result in results]
    supports=[result[1] for result in results]
    confidences=[result[2][0][3] for result in results]
    lifts=[result[2][0][3] for result in results]
    return list(zip(lhs,rhs,supports,confidences,lifts))
resultsinDataFrame=pd.DataFrame(inspect(results),columns=['Left Hand Side','Right Hand Side','Support','Confidence','Lift'])
resultsinDataFrame
```

```python
resultsinDataFrame.nlargest(n=10,columns='Lift')
```
